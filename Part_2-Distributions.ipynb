## Normal <a class="anchor" id="Normal"></a>
    - Also called Goussian distribution and bellcurve
    - Continuous Distribution
    - Mean = median = mode
    - Denoted by N~$(\mu,\sigma^2)$
    - As $\sigma$ increses the flatter the curve gets(fatter tail)

```python
# Genertate random number according to a normal distibution given mu and sigma
# distibution = np.random.normal(mu, sigma, 1000)

sigma_small = np.random.normal(0,2, 10000)
standard = np.random.normal(0,1, 10000)
sigma_large = np.random.normal(0,.6, 10000)
```

```python
# Generate Plot paramaters
fig, ax=plt.subplots(1,3,figsize=(10,3),sharey=True, sharex=True)

#Plot the dists
sns.distplot(sigma_small, ax=ax[0])
ax[0].set_title("Small sigma")
sns.distplot(standard, ax=ax[1])
ax[1].set_title("Standard")
sns.distplot(sigma_large, ax=ax[2],)
ax[2].set_title("Large Sigma")
```

#### Example: SAT scores

```python
# Example: SAT scores
scores = np.random.normal(500,100,size=10000)
```

```python
plot = sns.distplot(scores)
plot.set_title("Sample of 10000 SAT scores")
```

## Uniform <a class="anchor" id="Uniform"></a>
    - Descrete Distribtion
    - Each number has an equal probobility

```python
# Genertate random number according to a uniform distibution given range
# distibution = np.random.normal(mu, sigma, 1000)

uniform1 = np.random.uniform(1,10,size=10)
uniform2 = np.random.uniform(-10,10,size=100)
uniform3 = np.random.uniform(0,100,size=100)
```

```python
# Generate Plot paramaters
fig, ax=plt.subplots(1,3,figsize=(10,3),sharey=True, sharex=True)

#Plot the dists
sns.distplot(uniform1, ax=ax[0])
ax[0].set_title("Range 1 to 10")
sns.distplot(uniform2, ax=ax[1])
ax[1].set_title("Range -1 to 10")
sns.distplot(uniform3, ax=ax[2],)
ax[2].set_title("Range 0 to 100")
```

#### Example: Rolling a die

```python
# Exmple: Rolling a die
rolls = np.random.uniform(1,6,1000)
```

```python
plot = sns.distplot(rolls)
plot.set_title("1000 rolls of a die")
```

## Binomial <a class="anchor" id="Binomial"</a>
- Descrete variant of normal distribution

#### Example: Rolling 2 dice

```python
# Example: Rolling 2 dice
die1 = np.random.uniform(1,6,1000)
die2 = np.random.uniform(1,6,1000)
total = die1+die2
```

```python
plot = sns.distplot(total,bins=12)
plot.set_title("1000 rolls of two die")
```
